```python
import pandas as pd
import os

p65 = pd.read_excel(os.path.join("Vester.xlsx"),
                          index_col=0)
```

```python
p65.head()
```

```python
import networkx as nx
mapa = nx.from_pandas_adjacency(p65)
```

```python
#nodes
len(mapa)
```

```python
#edges
mapa.size()
```

```python
# plot
nx.draw_random(mapa,
                node_color='yellow',
                edge_color='blue',
                with_labels=True,
                font_size=8)
```

```python
mapa.remove_edges_from(nx.selfloop_edges(mapa))
```

```python
# re plot
nx.draw_random(mapa,
                node_color='yellow',
                edge_color='lightblue',
                with_labels=True,
                font_size=8)
```

```python
nx.draw_circular(mapa,
        with_labels=True, 
                 node_color='lightblue',
                edge_color='green',
                 font_size=10)
```

```python
nx.write_graphml(mapa, "Vesterp65Final.graphml",encoding='utf-8')
```
